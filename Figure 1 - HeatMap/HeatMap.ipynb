# **Importing Required Libraries**

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
```

# **Creating the `data` matrix**

```python
data = np.zeros((256, 256))
lim = 256

for cr in range(256):
  for cr1 in range(cr, 256):
    data[cr1, cr] = ((cr + 1) ** 2 + cr1 - cr + (lim - cr1) ** 2) / (lim ** 2)
```

# **Plotting complete HeatMap for `data`**

```python
plt.figure(figsize=(7.5, 6), dpi=100)

plot = sns.heatmap(data, cmap='hot')
plot.invert_yaxis()

plt.xlabel("C(r)")
plt.ylabel("C(r + 1)")
plt.xticks(list(range(0, 210, 50))+[255], list(map(str, range(0, 210, 50)))+['255'], rotation='horizontal')
plt.yticks(list(range(0, 210, 50))+[255], list(map(str, range(0, 210, 50)))+['255'], rotation='horizontal')
plt.figtext(0.46, 0.01, 'Probability map for P{y(n) = y(n+1)}', horizontalalignment='center', fontsize=16)
plt.show()

plot.get_figure().savefig('HeatMap_Actual.png')
```
