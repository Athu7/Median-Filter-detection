```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
files = ['/content/drive/MyDrive/MF_Detection_Results/original_result.txt',
         '/content/drive/MyDrive/MF_Detection_Results/median_3x3_result.txt',
         '/content/drive/MyDrive/MF_Detection_Results/median_5x5_result.txt',
         '/content/drive/MyDrive/MF_Detection_Results/median_7x7_result.txt',
         '/content/drive/MyDrive/MF_Detection_Results/median_9x9_result.txt']

res = []

for f in files:
  temp_file = open(f, 'r')
  values = np.array(eval(temp_file.read()))
  arr = []
  for i in range(201):
    x = np.sum(values > i / 200)
    arr.append(x / 1338)
  res.append(arr)
```

```python
plt.figure(figsize=(8, 6), dpi=100)
text = '''ROC Curve for all the classification between original
images and their median-filtered versions'''

plt.plot(res[0], res[1], '-', c='r', label='w = 3x3')
plt.plot(res[0], res[2], '--.', c='lime', markersize=5, label='w = 5x5')
plt.plot(res[0], res[3], ':', c='b', label='w = 7x7')
plt.plot(res[0], res[4], '-+', c='m', markersize=6, label='w = 9x9')
plt.xlim(-0.005, 0.14)
plt.ylim(0.7, 1.015)
plt.xlabel('FP', fontsize=16)
plt.ylabel('TP', fontsize=16)
plt.legend(prop={"size":15})
plt.figtext(0.12, -0.05, text, horizontalalignment='left', fontsize=16)
plt.savefig('roc.png')
plt.show()
```
